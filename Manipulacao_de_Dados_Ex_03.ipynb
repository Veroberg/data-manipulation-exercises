````python
import pandas as pd

users = pd.read_csv("/content/users_dataset.txt", sep='|')
users.head()
````

#### Tarefa 2. Descubra qual a idade média por ocupação.

*Dica: use a função groupby do Pandas na coluna occupation*

````python
media = users.groupby('occupation').mean()
display(media)
````

#### Tarefa 4. Para cada combinação de ocupação e sexo, calcule a idade média.

Exemplo:

```
occupation     gender
administrator  F         40.638889
               M         37.162791
artist         F         30.307692
               M         32.333333
```

````python
users_media = users.groupby(['occupation', 'gender'], as_index=False)['age'].mean()
display(users_media)
````

#### Tarefa 5. Para cada ocupação calcule a porcentagem de homens e mulheres.
